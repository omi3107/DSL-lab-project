### 1. Evaluate ASR (Whisper) using WER

```python
import evaluate
wer_metric = evaluate.load("wer")

references = ["the quick brown fox jumps over the lazy dog"]
predictions = ["the quick brown fox jumped over the lazy dog"] # "jumped" instead of "jumps"

wer_score = wer_metric.compute(predictions=predictions, references=references)
print(f"Word Error Rate (WER): {wer_score * 100:.2f}%")
```

### 2. Evaluate Summarization (BART) using ROUGE

```python
rouge_metric = evaluate.load("rouge")

ref_summaries = ["John assigned database migration to the backend team."]
pred_summaries = ["John will ask the backend team to migrate the database."]

rouge_scores = rouge_metric.compute(predictions=pred_summaries, references=ref_summaries)
print("ROUGE Scores:")
for k, v in rouge_scores.items():
    print(f"{k}: {v * 100:.2f}")
```

### 3. Evaluate NER (BERT) using Precision, Recall, and F1

```python
seqeval_metric = evaluate.load("seqeval")

labels = [["O", "B-PER", "O", "B-TASK", "I-TASK", "O", "B-DATE"]]
preds = [["O", "B-PER", "O", "B-TASK", "O", "O", "B-DATE"]]

results = seqeval_metric.compute(predictions=preds, references=labels)
print("NER Evaluation:")
print(f"Overall F1: {results['overall_f1']:.4f}")
print(f"Overall Precision: {results['overall_precision']:.4f}")
print(f"Overall Recall: {results['overall_recall']:.4f}")
```
